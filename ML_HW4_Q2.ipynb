```
import numpy as np
import matplotlib.pyplot as plt

cov = [[0.02, 0], [0, 1]]

mean_1 = [0.5, 0.5]
cluster_1 = np.random.multivariate_normal(mean_1, cov, 100)

mean_2 = [-0.5, -0.5]
cluster_2 = np.random.multivariate_normal(mean_2, cov, 100)
print('| Part (a) |')
plt.scatter(*zip(*cluster_2))
plt.scatter(*zip(*cluster_1))
plt.show()
```

```
# From class, assuming b = 0 (w is homogeneous and passes through the origin)
w = np.zeros(2)
w_last = np.zeros(2)
T = 100
alpha = 1

epoch = 0

while epoch < T:
  for sample in cluster_1:
    if np.dot(w, sample) < 0:
      w = w + 1*sample
  
  for sample in cluster_2:
    if np.dot(w, sample) >= 0:
      w = w + -1*sample

  if np.array_equal(w_last, w):
    break
  else:
    w_last[0] = w[0]
    w_last[1] = w[1]
    epoch += 1

print('| Part (b) |')
print('\tCONVERGED' if epoch < 100 else '|DID NOT CONVERGE|')
print('\tEpochs = ', epoch, '| weights = ', w)
plt.scatter(*zip(*cluster_2))
plt.scatter(*zip(*cluster_1))

line_x = np.linspace(-0.15,0.15,2)
line_y = ((-1*w[0])/w[1])*line_x + 0

plt.plot(line_x, line_y, linestyle='--', color='r')
plt.show()
```

```
# Part (c) with overlapping classes

cov = [[0.1, 0], [0, 10]]

mean_1 = [0.5, 0.5]
cluster_1 = np.random.multivariate_normal(mean_1, cov, 100)

mean_2 = [-0.5, -0.5]
cluster_2 = np.random.multivariate_normal(mean_2, cov, 100)

print('| Part (c) |')
plt.scatter(*zip(*cluster_2))
plt.scatter(*zip(*cluster_1))
plt.show()
```

```
# From class, assuming b = 0 (w is homogeneous and passes through the origin)
w = np.zeros(2)
w_last = np.zeros(2)
T = 100
alpha = 1

epoch = 0

while epoch < T:
  for sample in cluster_1:
    if np.dot(w, sample) < 0:
      w = w + 1*sample
  
  for sample in cluster_2:
    if np.dot(w, sample) >= 0:
      w = w + -1*sample

  if np.array_equal(w_last, w):
    break
  else:
    w_last[0] = w[0]
    w_last[1] = w[1]
    epoch += 1

print('| Part (c) |')
print('\tCONVERGED' if epoch < 100 else '|DID NOT CONVERGE|')
print('\tEpochs = ', epoch, '| weights = ', w)
plt.scatter(*zip(*cluster_2))
plt.scatter(*zip(*cluster_1))

line_x = np.linspace(-1,1,2)
line_y = ((-1*w[0])/w[1])*line_x + 0

plt.plot(line_x, line_y, linestyle='--', color='r')
plt.show()
```

**Part (b)**<br />
  The implementation **converges** within 1 to 3 epochs, depending on the generated distribution.

**Part (c)**<br />
  The implementation **does not converge** even after 100 epochs. The classification regions, shown above, are not perfect but separate most of the samples correctly.
